## Session #7: The Expectation Maximization algorithm for Gaussian Mixture models


We are given a dataset $D=\{\textbf{x}_1,\textbf{x}_2,...,\textbf{x}_N\}$, where $\textbf{x}_i$ is a $d$-dimentional vector.

* Assume that the data points are generated in an i.i.d pattern from an underlying density function $p(\textbf{x})$.
* We further assume that $p(\textbf{x})$ is defined as a mixture of $K$ Gaussian components:

$$p(\textbf{x}|\theta)=\sum_z p(\textbf{z})p(\textbf{x}|\textbf{z},\theta)=\sum_{k=1}^K\pi_k N(\textbf{x}|\mu_k,\Sigma_k)$$

- $\textbf{z}$ is a $K$-dimentional binary variable. The values of $z_k$ satisfy $z_k\in\{0,1\}$ and $\sum_k z_k=1$.
- The $\pi_k=p(z_k=1)$ are rhe mixture weights (or mixing coefficients), representing the probability that a randomly selected $\textbf{x}$ was generated by component $k$, where $\sum_k\pi_k=1$



### EM for Gaussian Mixtures

* Initialize the mean $\mu_k$, covariances $\Sigma_k$ and mixing coefficients $\pi_k$, and evaluate the initial value of the log likelihood.
* **E-step.** Compute the membership weights 

$$w_{nk}=\frac{\pi_k N(\textbf{x}_n|\mu_k,\Sigma_k)}{\sum_{j=1}^K \pi_jN(\textbf{x}_n|\mu_j,\Sigma_j)}$$

* **M-step.** Reestimate the parameters using the current membership weights

$$\mu_k^{new}=\frac{1}{N_k}\sum_{n=1}^N w_{nk} \textbf{x}_n$$

$$\Sigma_k^{new}=\frac{1}{N_k}\sum_{n=1}^N w_{nk} (\textbf{x}_n-\mu_k^{new})(\textbf{x}_n-\mu_k^{new})^T$$

$$\pi_k^{new}=\frac{N_k}{N}$$ where $N_k=\sum_{n=1}^Nw_{nk}$

* Evaluate the log likelihood

$$ln\quad p(\textbf{X}|\mu,\Sigma,\pi)=\sum_{n=1}^N ln \{\sum_{k=1}^K \pi_k N(\textbf{x}_n|\mu_k,\Sigma_k)\}$$

### Initialization

The EM algorithm can be started by either initializing the algorithm by a set of parameters and then conducting an E-step, or by starting with a set of initial membership weights and then doing a M-step.

The initial parameters or weights can be chosen randomly or could be chosen via some heuristic method (such as by using the K-means to cluster the data and then defining weights based on K-means membership).

Import packages

```python
%matplotlib inline

import numpy as np

from matplotlib import pyplot as plt
from sklearn import metrics, datasets
from sklearn.preprocessing import scale, normalize

from scipy.stats import multivariate_normal as mvn
```

Generate some data from a Gaussian mixture model

```python
np.random.seed(0)

# Define the parameters of a mixture of 3 Gaussian
mus = np.array([[0,4], [-2,0], [6,-2]])
sigmas = np.array([ [[3, 0], [0, 0.5]], [[1,0],[0,2]], [[3,0],[0,1]] ])
pis = np.array([0.3, 0.4, 0.3])

# Generate some data from the above mixture
samples=1000
dim=2
data = []
label= []
k=0
for pi, mu, sigma in zip(pis, mus, sigmas):
    components_samples=np.int(pi*samples)
    data.append(pi*mvn.rvs(mu, sigma,components_samples))
    label.append([k]*(components_samples))
    k = k+1

data=np.asarray(data)
label=np.asarray(label)
data=np.concatenate((data[0], data[1], data[2]), axis=0)
label=np.concatenate((label[0], label[1], label[2]), axis=0)
```

Plot the data

```python
K = len(mus) # number of clusters
colors = ['navy', 'c', 'cornflowerblue']
plt.figure(figsize=(8, 6))

for k in range(K):
    plt.scatter(data[label == k, 0], data[label == k, 1], color = colors[k % len(colors)])
plt.title('Data from Gaussian Mixture Model')
plt.show()
```

Implementation of the EM algorithm

```python
# The EM algorithm
def em_gmm(xs, pis, mus, sigmas, ws, val, tol=0.000001, max_iter=1000): 
    # val indicates if E or M step is conducted in first iteration

    n, p = xs.shape  # Dimensions of data set
    k = K            # Number of clusters
    
    # Log likelihoods
    lls = []
    ll_old = 0
    
    for m in range(max_iter):

        # E-step
        if(val==0 or m>=1):
            # Membership weights
            ws = np.zeros((k, n))
            for j in range(k):
                for i in range(n):
                    ws[j, i] = pis[j] * mvn(mus[j], sigmas[j]).pdf(xs[i])
            ws = normalize(ws,norm='l1', axis=0)
        
        
        # M-step
        if(val==1 or m>=1):
            # Mixture weights
            pis = np.zeros(k)
            for j in range(k):
                for i in range(n):
                    pis[j] += ws[j, i]
            pis /= n
            
            # Means
            mus = np.zeros((k, p))
            for j in range(k):
                for i in range(n):
                    mus[j] += ws[j, i] * xs[i]
                mus[j] /= ws[j, :].sum()
            
            # Covariance matrix
            sigmas = np.zeros((k, p, p))
            for j in range(k):
                for i in range(n):
                    ys = np.reshape(xs[i]- mus[j], (2,1))
                    sigmas[j] += ws[j, i] * np.dot(ys, ys.T)
                sigmas[j] /= ws[j,:].sum()

        # Update Log Likelihoood
        ll_new = 0.0
        for i in range(n):
            s = 0
            for j in range(k):
                s += pis[j] * mvn(mus[j], sigmas[j]).pdf(xs[i])
            ll_new += np.log(s)

        # Convergence check
        if (np.abs(ll_new - ll_old) < tol and m > 1):
            break
        ll_old = ll_new
        lls.append(ll_new)

    return lls, ll_new, pis, mus, sigmas
```

Predicted labels from EM

```python
def predict(xs, pis, mus, sigmas):
    k = K
    n=samples
    ws = np.zeros((k, n))
    for j in range(k):
        for i in range(n):
            ws[j, i] = pis[j] * mvn(mus[j], sigmas[j]).pdf(xs[i])
    ws = normalize(ws,norm='l1', axis=0)
    estimated_label = np.zeros(n)
    for i in range(n):
        estimated_label[i]=np.argmax(ws[:,i])
    estimated_label=estimated_label.T
    return(estimated_label)
```

Functions for plotting the data and ellipses

```python
# Plotting functions
from matplotlib.patches import Ellipse
    
def plot_ellipse(pos, cov, nstd=2, ax=None, **kwargs):
    def eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]

    if ax is None:
        ax = plt.gca()

    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))

    width, height = 2 * nstd * np.sqrt(abs(vals))
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, **kwargs)

    ax.add_artist(ellip)
    return ellip    
    
def show(X, y, mus, sigmas, colors_ell):
    plt.cla()
    K = len(mus) # number of clusters
    colors = ['navy', 'c', 'cornflowerblue']
    
    for k in range(K):
        plt.plot(X[y==k,0], X[y==k,1], '.', color = colors[k % len(colors)])
        plot_ellipse(mus[k], sigmas[k],  alpha=0.7, color = colors_ell[k % len(colors)])  
```

#### 1. Initializing the algorithm by a set of parameters and conduct an E-step
Random parameter initialization

```python
np.random.seed(0)

# Mixture weights
pis = np.random.random(K)
pis /= pis.sum() 

# Means
mus = np.random.random((K,dim))

# Covariance matrix
sigmas = np.array([np.eye(dim)] * K)

# Membership weights (not used here)
ws = np.zeros((K, samples))
```

```python
# Run the EM algorithm using initial parameters
lls, ll, pis, mus, sigmas = em_gmm(data, pis, mus, sigmas, ws, 0)
```

```python
# Plotting the results
colors_ell = ['c', 'cornflowerblue','navy']
fig = plt.figure(figsize = (16, 6))
fig.add_subplot(121)
show(data, label, mus, sigmas, colors_ell)
plt.axis([-3,4,-2.5,2.5])
plt.title('EM for Gaussian Mixtures')
fig.add_subplot(122)
plt.plot(lls)
plt.title('Log Likelihood Plot')
plt.xlabel('Iterations')
plt.ylabel('Log Likelihood')
plt.show()
```

```python
# Evaluating the classification
label_pred = predict(data, pis, mus, sigmas)

print('Homogenity score: %0.4f' % metrics.homogeneity_score(label,label_pred))
print('Completeness score: %0.4f' % metrics.completeness_score(label,label_pred))
print('VMeasure score: %0.4f' % metrics.v_measure_score(label,label_pred))
print('Adjusted random score: %0.4f' % metrics.adjusted_rand_score(label,label_pred))
print('Adjusted mutual info score: %0.4f' % metrics.adjusted_mutual_info_score(label,label_pred))
print('Silhouette score: %0.4f' % metrics.silhouette_score(data,label_pred))
```

#### 2. Start with a set of initial membership weights and conduct a M-step
Random parameter initialization

```python
# Membership weights
ws = np.zeros((K, samples))
for i in range(K):
    for j in range(samples):
        ws[i,j]=np.random.rand()
ws=normalize(ws,norm='l1',axis=0)
```

```python
# Run the EM algorithm using initial parameters
lls, ll, pis, mus, sigmas = em_gmm(data, pis, mus, sigmas, ws, 1)
```

```python
# Plotting the results
colors_ell = ['cornflowerblue','navy','c']
fig = plt.figure(figsize = (16, 6))
fig.add_subplot(121)
show(data, label, mus, sigmas, colors_ell)
plt.axis([-3,4,-2.5,2.5])
plt.title('EM for Gaussian Mixtures')
fig.add_subplot(122)
plt.plot(lls)
plt.title('Log Likelihood Plot')
plt.xlabel('Iterations')
plt.ylabel('Log Likelihood')
plt.show()
```

```python
# Evaluating the classification
label_pred = predict(data, pis, mus, sigmas)

print('Homogenity score: %0.4f' % metrics.homogeneity_score(label,label_pred))
print('Completeness score: %0.4f' % metrics.completeness_score(label,label_pred))
print('VMeasure score: %0.4f' % metrics.v_measure_score(label,label_pred))
print('Adjusted random score: %0.4f' % metrics.adjusted_rand_score(label,label_pred))
print('Adjusted mutual info score: %0.4f' % metrics.adjusted_mutual_info_score(label,label_pred))
print('Silhouette score: %0.4f' % metrics.silhouette_score(data,label_pred))
```

When comparing the two likelihood functions we observe that we obtain convergence in the log likelihood after approximately the same number of observations. By initializing the EM algorithm with the means, covariance matrix and mixture weights and conducting the E-step we see that the log likelihood value show a steaper increase in the first iterations than for the M-step. The increase in the log likelihood for the E-step initialization slowly decrease. For the M-step we observe a steap increase in the log likelihood after a few iterations. Then the increase flattens out before it increase again and reach the defined level for convergence.

#### Alternative initialization

```python
import random
import numpy.linalg as la

# Means
r = random.Random(1)
mus=list(map(lambda _: r.choice(data), range(K)))
mus=np.asarray(mus)

# Convariance matrix
clusters = [[] for i in range(K)]
for d in data:
    i = np.argmin([la.norm(d - m) for m in mus])
    clusters[i].append(d)

for i in range(K):
    sigmas[i] = np.cov(clusters[i], rowvar=0)

# Mixture weights
pis=[]
for i in range(K):
    pis.append(len(clusters[i])/len(data))
```

```python
# Run the EM algorithm using initial parameters
lls, ll, pis, mus, sigmas = em_gmm(data, pis, mus, sigmas, ws, 0)
```

```python
# Plotting the results
colors_ell = ['navy','c','cornflowerblue']
fig = plt.figure(figsize = (16, 6))
fig.add_subplot(121)
show(data, label, mus, sigmas, colors_ell)
plt.axis([-3,4,-2.5,2.5])
plt.title('EM for Gaussian Mixtures')
fig.add_subplot(122)
plt.plot(lls)
plt.title('Log Likelihood Plot')
plt.xlabel('Iterations')
plt.ylabel('Log Likelihood')
plt.show()
```

```python
# Evaluating the classification
label_pred = predict(data, pis, mus, sigmas)

print('Homogenity score: %0.4f' % metrics.homogeneity_score(label,label_pred))
print('Completeness score: %0.4f' % metrics.completeness_score(label,label_pred))
print('VMeasure score: %0.4f' % metrics.v_measure_score(label,label_pred))
print('Adjusted random score: %0.4f' % metrics.adjusted_rand_score(label,label_pred))
print('Adjusted mutual info score: %0.4f' % metrics.adjusted_mutual_info_score(label,label_pred))
print('Silhouette score: %0.4f' % metrics.silhouette_score(data,label_pred))
```
