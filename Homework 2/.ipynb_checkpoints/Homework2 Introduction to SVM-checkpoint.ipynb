### 1. Linear SVM

a) Generate a two-dimensional linearly separable dataset with 300 samples. We consider a two-class balanced classification problem. The samples from the first class are normally distributed with $\mu=0$ and $\sigma=1.5$, while the samples from the second class have $\mu=3$ and $\sigma=1.5$. 

```python
import numpy as np

# Create 300 separable points
rng = np.random.RandomState(0)
n1=150
n2=150
X=np.r_[1.5*rng.randn(n1,2),1.5*rng.randn(n2,2)+[3,3]]
y=[0]*n1 + [1]*n2
```

Use a Scatter plot to show the data in a two-dimensional space and give a different color to each class. 

```python
import matplotlib.pyplot as plt

# Scatter plot of matrix X
plt.figure(figsize=(8, 6))
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.title('Scatter plot linear')
plt.show()
```

b) Fit a linear SVC using part of the data in X as training set (X_train, y_train) and holding out part of the available data as a test set (X_test, y_test). 

```python
from sklearn import svm
from sklearn.cross_validation import train_test_split

# Create train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Fit linear SVC
lin_svc = svm.SVC(kernel='linear')
lin_svc.fit(X_train, y_train)
```

c) Get and plot the separating hyperplane and the parallels to the separating hyperplane that pass through the support vectors. How about the number of support vectors?

```python
# Get the separating hyperplane
w=lin_svc.coef_[0]
b= lin_svc.intercept_[0]
xx=np.linspace(-5,5)
yy=-b/w[1]-(w[0]/w[1])*xx

# Get the parallels to the separating hyperplane
yy_down = 1/w[1]-b/w[1]-(w[0]/w[1])*xx
yy_up = -1/w[1]-b/w[1]-(w[0]/w[1])*xx

# Find number of support vectors
n = len(lin_svc.support_vectors_)

# Plot the lines
plt.figure(figsize=(8, 6))
plt.plot(xx,yy,'k-')
plt.plot(xx,yy_down,'k--')
plt.plot(xx,yy_up,'k--')


# Plot the points
plt.scatter(lin_svc.support_vectors_[:,0],lin_svc.support_vectors_[:,1],s=80,facecolors='black')
plt.scatter(X[:,0],X[:,1], c=y, cmap=plt.cm.coolwarm)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.axis('tight')
plt.title('Linear SVC, C=%d SV=%d' % (lin_svc.C, n))
plt.show()
```

d) Compute the classification scores both on training and test sets.

```python
# The mean accuracy on training sets
print('Training error: %0.4f' % (1-lin_svc.score(X_train,y_train)))

# The mean accuracy on the given test data and labels
print('Test error: %0.4f' % (1-lin_svc.score(X_test,y_test)))
```

e) Fit the model for different values of the penalty parameter C (from very low values to very high ones). Study the influence of this parameter on the classification score and on the number of support vectors. Why the margin is larger for low values of C than it is for higher values?

```python
C_range = [0.001,0.01,0.1,1,10,1000]
error=[]
numSV=[]

# Set the figure size
plt.figure(figsize=(16, 16))

for i in range(0,len(C_range)):
    # Fit the linear SVC
    c=C_range[i]
    lin_svc=svm.SVC(kernel='linear',C=c)
    lin_svc.fit(X_train,y_train)
    
    # Find the error
    error.append(1-lin_svc.score(X_test,y_test))

    # Find number of support vectors
    n = len(lin_svc.support_vectors_)
    numSV.append(n)

    # Get the separating hyperplane
    w=lin_svc.coef_[0]
    b= lin_svc.intercept_[0]
    xx=np.linspace(-5,5)
    yy=-b/w[1]-(w[0]/w[1])*xx

    # Get the parallels to the separating hyperplane
    yy_down = 1/w[1]-b/w[1]-(w[0]/w[1])*xx
    yy_up = -1/w[1]-b/w[1]-(w[0]/w[1])*xx

    # Plot the lines
    plt.subplot(3,2,i+1) 
    plt.plot(xx,yy,'k-')
    plt.plot(xx,yy_down,'k--')
    plt.plot(xx,yy_up,'k--')
    
    # Plot the points
    plt.scatter(lin_svc.support_vectors_[:,0],lin_svc.support_vectors_[:,1],s=80,facecolors='black')
    plt.scatter(X[:,0],X[:,1], c=y, cmap=plt.cm.coolwarm)
    plt.xlabel('X_1')
    plt.ylabel('X_2')
    plt.axis('tight')
    plt.title('Linear SVC, C= %s SV= %d' % (str(c),n))

# Set figure size
plt.figure(figsize=(16, 5))

# Error plot for different c parameters
error_plot = plt.subplot(121)
error_plot.plot(C_range,error)
plt.xlabel('C')
plt.ylabel('Test error')
plt.xscale('log')
plt.title('Plot of test error')

# Plot of number of support vectors for different c parameters
numSV_plot = plt.subplot(122)
numSV_plot.plot(C_range,numSV)
plt.xlabel('C')
plt.ylabel('Number of SV')
plt.xscale('log')
plt.title('Number of support vectors')

plt.show()
```

The C parameter is telling the SVM optimizer how much weight you put on avoiding misclassifying. A high C value will result in a smaller margin because it will lead to more points being classified correctly. Conversely, a low value of C will try to fit a separating hyperplane margin because the misclassification of points is of less importance.

### 2. Nonlinear SVM

a) Generate a two-dimensional non-linearly separable dataset with 300 samples. For that, start from a two-dimensional array (X) of normally distributed data and get a target vector (y) using the logical XOR operator. 

```python
# Create 300 separable points
rng = np.random.RandomState(0)
X = rng.randn(300, 2)
y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0)
```

b) Visualize the data using a scatter plot and give a different color to each class. 

```python
# Scatter plot of matrix X
plt.figure(figsize=(8, 6))
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm)
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.title('Scatter plot non-linear')
plt.show()
```

c) Fit a RBF kernel SVC using the default parameters. Plot the decision function for each datapoint (you can take as baseline the function plot_svm_nonlinear.py from <a href = http://scikit-learn.org/stable/auto_examples/svm/plot_svm_nonlinear.html> scikit-learn.org/stable/auto_examples/svm/plot_svm_nonlinear.html</a>)

```python
# Create grid
xx, yy = np.meshgrid(np.linspace(-3.5, 3, 300),np.linspace(-3.5, 3, 300))

# Fit the non-linear SVC
rbf_svc = svm.SVC(kernel='rbf')
rbf_svc.fit(X, y)

# Set figure size
plt.figure(figsize=(8, 6))

# Plot the decision function for each datapoint on the grid
Z = rbf_svc.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.imshow(Z, interpolation='nearest',extent=(xx.min(), xx.max(), yy.min(), yy.max()), 
           aspect='auto',origin='lower', cmap=plt.cm.RdBu)
contours = plt.contour(xx, yy, Z, levels=[0], colors='#000066', linewidths=2, linetypes='--')

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
plt.xticks(())
plt.yticks(())
plt.axis('tight')
plt.title('Non-linear SVC, C=10^%d' % (np.log10(rbf_svc.C)))

plt.show()
```

d) Study how the kernel parameter $\gamma$ and the penalty parameter C affect the shape of the decision function. Tune these parameters to obtain:
    * A decision boundary almost linear
    * A very sharp decision boundary
    * A curved, smooth decision boundary

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

# Scale the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Use gridsearch to investigate different c and gamma parameters
C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
grid = GridSearchCV(svm.SVC(kernel='rbf'), param_grid=param_grid, cv=cv)
grid.fit(X, y)

# Obtain the best parameters
print('The best parameters are %s with a score of %0.2f'% (grid.best_params_, grid.best_score_))
```

```python
C_2d_range = [1e-1, 1, 1e1]
gamma_2d_range = [1e-3, 1e-2, 1e-1, 1, 1e1, 1e2, 1e3]
classifiers = []
for gamma in gamma_2d_range:
    for c in C_2d_range:
        clf = svm.SVC(kernel='rbf',C=c, gamma=gamma)
        clf.fit(X, y)
        classifiers.append((gamma, c, clf))

# Set figure size
plt.figure(figsize=(16, 24))

# Create grid
xx, yy = np.meshgrid(np.linspace(-3.5, 3, 300), np.linspace(-3.5, 3, 300))

for (k, (gamma, c, clf)) in enumerate(classifiers):
    # Evaluate decision function in a grid
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot the decision function for these parameters
    plt.subplot(len(gamma_2d_range), len(C_2d_range), k + 1)
    plt.title("Non-linear SVC, C=10^%d, gamma=10^%d" % (np.log10(c), np.log10(gamma)))
    
    #plt.pcolormesh(xx, yy, -Z, cmap=plt.cm.RdBu)
    plt.imshow(Z, interpolation='nearest',extent=(xx.min(), xx.max(), yy.min(), yy.max()), aspect='auto',
               origin='lower', cmap=plt.cm.RdBu)
    contours = plt.contour(xx, yy, Z, levels=[0], colors='#000066', linewidths=2, linetypes='--')
    
    # Plot the parameter's effect on decision function
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
    plt.xticks(())
    plt.yticks(())
    plt.axis('tight')

plt.show()
```

e) Fit a Polynomial kernel SVC with degreees 1, 2, 3, 4. Interprete the results.

```python
h = .02  # step size in the mesh
C = 1.0  # SVM regularization parameter

# Create a grid
xx, yy = np.meshgrid(np.linspace(-3, 3, 200), np.linspace(-3, 3, 200))


# Title for the plots
titles = ['Polynomial SVC with degree 1',
          'Polynomial SVC with degree 2',
          'Polynomial SVC with degree 3',
          'Polynomial SVC with degree 4'
         ]

# Set figure size
plt.figure(figsize=(16, 12))

for i in range(1,5,1):
    # Fit polynomial SVC
    poly_svc = svm.SVC(kernel='poly', degree=i, C=C).fit(X, y)
    
    # Plot decision function
    plt.subplot(2, 2, i)
    Z = poly_svc.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.7, linetypes='--')
    plt.contour(xx, yy, Z, colors='#000099', linewidth=.5)

    # Plot the points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i-1])

plt.show()
```

From the plots it is clear that the second degree polynomial kernel classifier is the most accurate.
