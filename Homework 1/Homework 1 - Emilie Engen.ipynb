Create a two-dimensional array (X) of size (200x2) composed by normal random numbers

```python
# Import numpy package
import numpy as np

rnd = np.random.RandomState(0) # Method for generating random numbers
X = rnd.randn(200, 2) # Return matrix of random standard normal numbers

print ('\nX matrix: \n' + str(X)) # Print matrix
print ('\nDimensions of matrix X: ' + str(X.shape)) # Get shape of matrix
```

Use a Scatter plot to show the data in a two-dimensional space

```python
# Import plot packages
import matplotlib.pyplot as plt

# Scatter plot of matrix X
plt.scatter(X[:,0],X[:,1])
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.show()
```

Transform the element of X into boolean values according to the following statement: Positive elements correspond to True and negative ones to False

```python
# Transform X matrix to boolean values
X_bool = X >= 0

# Print matrix X_bool
print ('\nBoolean X matrix: \n' + str(X_bool))
```

Generate a target variable Y using the logical XOR operator applied to the boolean elements of X

```python
# Compute target variable Y = X_1 XOR X_2
Y = np.logical_xor(X_bool[:,0],X_bool[:,1])

# Print Y matrix
print ('\nMatrix of target variable Y: \n' + str(Y))
```
